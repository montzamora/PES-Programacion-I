```python
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # for 3D plotting

x = np.arange(150)
f = lambda x: x**3
y = f(x)
```

```python
# create a figure/canvas of desired size
graf = plt.figure(figsize=(8, 6))

# plot values; with a color `red`
plt.plot(x, y, 'b',color='green')

# give labels to the axes
plt.xlabel('x')
plt.ylabel('y')

# give a title to the plot
plt.title(r"Plot of $y=x^3$")

plt.savefig("./graf.pdf", format='pdf', dpi=80)

plt.show()
```

```python
df = pd.DataFrame({'X':x,'Y':y})
```

```python
grafica = df.plot.scatter(x='X', y='Y')
grafica.get_figure().savefig('figura.pdf',dpi=80)
```

Resuelva las siguientes preguntas utilizando las funciones de pandas. Ayúdese con la hoja de trucos.

```python
# Tu código acá
male = df[df['Sex'] == 'male']
female = df[df['Sex'] == 'female']

print("La edad media de los hombres es :",male['Age'].mean())
print("La edad media de las mujeres es :",female['Age'].mean())
```

- Calcula la edad promedio para cada sexo. Ahora utiliza el método `groupby`.

- Calcule esta relación de supervivencia para todos los pasajeros menores de 25 años (recuerde: filtrado / indexación booleana).

```python
# Tu código acá
menores_25 = df[df['Age'] < 25]['Survived'].sum()/df[df['Age'] < 25]['Survived'].count()*100
print('La tasa de supervivencia es %0.3f por ciento.' % menores_25)
```

- ¿Cuál es la diferencia en la proporción de supervivencia entre sexos?

```python
# Tu código acá
df = pd.read_csv("data/titanic.csv")
male = df[df['Sex'] == 'male']
female = df[df['Sex'] == 'female']
sobrev_male = (male['Survived'].sum()/male['Survived'].count())*100
sobre_female = (female['Survived'].sum()/female['Survived'].count())*100
print("Sobrevivencia hombres:", sobrev_male)
print("Sobrevivencia mujeres:", sobre_female)
print("La diferencia es:", (sobre_female-sobrev_male))
```

- ¿O cómo difiere la proporción de sobrevivientes entre las diferentes clases de pasajeros? Haz una gráfica de barras visualizando la relación de supervivencia para las 3 clases.

```python
# Tu código acá
clase = df.groupby('Pclass')['Survived'].sum()/df.groupby('Pclass')['Survived'].count()
clase
```

```python
clase.plot(kind='bar', color='brown')
```

- Haga una gráfica de barras para visualizar la tarifa promedio pagada por los pasajeros de acuerdo a su edad. Para esto:
    - Primero, separamos a los pasajeros por rango, utilizando la función `pd.cut` y agregamos esta serie al DataFrame.
    - Debemos agrupar por esta columna y calcular el promedio de las tarifas.
    - Por último, agregamos `plot(kind='bar')` para obtener la gráfica de barras.

```python
df['AgeClass'] = pd.cut(df['Age'], bins=np.arange(0,90,10))
# Tu código acá
```

```python
promedio = df.groupby(by='AgeClass')['Fare'].mean()
promedio.plot(kind='bar',color='black')
```
